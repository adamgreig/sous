Sous Vide Kettle Controller===========================Problem Statement-----------------We have a standard electric kettle. At first glance it holds 1.7L water and probably has an approximately 3kW heating element.We wish to convert it to a sous vide cooker, which will require holding a temperature between 50-80C to within 0.5C for anindefinite period of time, from hours to days.Kettle Model------------The kettle state is the temperature of the water, $y$.The heater increases the temperature of the water as $x\dot{Q}=mc\dot{y}$ where:$x$ is the control input bounded by $(0,1)$, $\dot{Q}$ is the heater maximum power,$m$ is the mass of water (or, well, total contents of kettle, assumed to be similar)and $c$ is the specific heat capacity of those contents, $c \approx 4.1813$.The kettle also suffers from heat loss as $\dot{y}=-ay$ where $a$ is a parameter to be found,assuming the ambient temperature is roughly constant.Overall then $\dot{y} = \frac{x\dot{Q}}{mc} - ay$.System Indentification----------------------1.7L of water was boiled in the kettle and left to cool to 50C while temperature measurements were taken.1.7L water $\Rightarrow m=1700$.First we fit to the heating portion of the data. Assuming negligable heat loss during this period,and with the kettle on full power, $\dot{y} = \frac{\dot{Q}}{mc} \Rightarrow y=\frac{\dot{Q}}{mc}t + b$

```python
import numpy as npimport matplotlib.pyplot as pltfrom scipy import optimizeheating_times = np.array([0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300])heating_temps = np.array([7, 10, 20, 30, 40, 50, 59, 69, 78, 87, 96])def linear(x, a, b):    return a*x + bpopt, pcov = optimize.curve_fit(linear, heating_times, heating_temps)print "qdot / mc =", popt[0]print "qdot =", popt[0] * 1700 * 4.1813print "variance in qdot estimate:", pcov[0][0] * 1700 * 4.1813plt.scatter(heating_times, heating_temps, label="Experimental Data")plt.plot(heating_times, linear(heating_times, *popt), label="Line Fit")_ = plt.xlabel("Time (s)")_ = plt.ylabel("Temperature (C)")_ = plt.legend(loc=2)
```

We see that $\dot{Q} = 2205$ to some reasonable variance. Perhaps the kettle has a 2.2kW heater.Next we fit to the heat loss section of the data to determine $a$.$\dot{y} = 0 - ay \Rightarrow y = ce^{-at}$

```python
cooling_times = np.array([360, 540, 840, 1140, 1440, 1740, 2040, 2340, 2640, 2940, 3240, 3540, 3840, 4200, 4440, 4740, 5040, 5340, 5640, 5940, 6240, 6540, 6840, 7140]) - 360cooling_temps = np.array([102, 99, 93, 89, 85, 82, 79, 77, 74, 72, 69, 68, 66, 63, 62, 60, 59, 57, 56, 54, 53, 52, 51, 50])def exponential(x, b, a):    return b*np.exp(-a*x)popt, pcov = optimize.curve_fit(exponential, cooling_times, cooling_temps, np.array([96, 0.0001]))print "a =", popt[1]print "variance in a estimate:", pcov[1][1]plt.scatter(cooling_times, cooling_temps, label="Experimental Data")plt.plot(cooling_times, exponential(cooling_times, *popt), label="Exponential Fit")_ = plt.xlabel("Time (s)")_ = plt.ylabel("Temperature (C)")_ = plt.legend()
```

We obtain $a=0.001072$ though the fit is perhaps less convincing. Never mind, that's what feedback is for!So then:$\dot{y} + 0.001072y = 0.3103x$Taking Laplace transforms,$sY(s) + 0.001072Y(s) = 0.3103X(s)$$Y(s)(s + 0.001072) = 0.3103X(s)$$\frac{Y(s)}{X(s)} = \frac{0.3103}{s + 0.001072} = G(s)$

```python
import controlG = control.tf(0.3103, [1, 0.001072])_ = control.bode(G)
```

Controlling The Kettle----------------------Clearly the kettle has phase lag (unsurprisingly) so we use a phase lag compensator.$K=k\frac{s+\alpha}{s+\beta}$

```python
k = 1.0alpha = 0.5beta = 0.5K = k * control.tf([1, alpha], [1, beta])# haha let's just use K=1K = 1.0GK = control.series(G, K)CL = GK / (1+GK)[T, yout, xout] = control.forced_response(CL, np.linspace(0, 500, 5000), U=60, X0=7)plt.subplot(211)plt.plot(T, yout)plt.grid()plt.ylim((0, 70))plt.xlabel("Time (s)")plt.ylabel("Temperature (C)")plt.title("Controlled kettle moving to and holding 60C")plt.subplot(212)plt.plot(T, yout)plt.grid()plt.ylim((55, 65))plt.xlabel("Time (s)")plt.ylabel("Temperature (C)")print "Final value:", yout[-1]
```

Sous Vide Cooker Control========================Setup: 2L water in a kettle with a 3kW heater that is under variable control. The specific heat capacity of water is 4.1813J/gK, we have 2000g of water and 3000W of power.System Equation: $ \dot{Q} = m c \frac{dT}{dt} $We have: $y=T$, $x=\dot{Q}/3000$Giving: $\dot{y} = 0.35874 x$Take Laplace transform: $sY(s) - y(0) = \alpha X(s) \Rightarrow \frac{Y(s)}{X(s)} = G(s) = \frac{0.35874}{s}$

```python
from control import *import numpy as npimport matplotlib.pyplot as pltG = tf(0.35874, [1, 0])_ = bode(G)
```

Clearly we have a lag so use a phase lag compensator.$K(s) = k\frac{s + \alpha}{s + \beta}, k=1, \alpha=0.3, \beta=0.3$For the closed loop:        r        e           x          y     ------>O------->| K |----->| G |--------->           -|                           |            -----------------------------$y = GK(r - y)$$y(1+GK) = GKr$$y/r = GK/(1+GK)$

```python
alpha = 0.5beta = 0.05K = tf([1, 0.3], [1, 0.3])GK = series(G, K)CL = GK / (1 + GK)tt, yy, xx = forced_response(CL, T=np.linspace(0, 300, 3000), U=56.0, X0=15.0)plt.plot(tt, yy)title("Cooker Profile, first five minutes, 15C to 56C")xlabel("Time (s)")ylabel("Temperature (C)")grid()
```
